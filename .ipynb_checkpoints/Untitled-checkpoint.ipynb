```python
import warnings
warnings.filterwarnings('ignore')

from IPython.display import display, HTML

from tabulate import tabulate

from itertools import product

import matplotlib.ticker as mtick

import numpy as np
import pandas as pd
import pandas_profiling
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.tree import plot_tree

#estimators
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
#model metrics

from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

%matplotlib inline

def get_bins(data,ind_vars,dep_var,bins):
    bins_str=''
    bin_cts=''
    for key,value in sorted(bins.items()):
        min_value = value[1] #data[key].min()-1
        max_value = value[2] #data[key].max()
        size = np.floor(np.log10(max_value))
        max_value = 10*size*np.ceil(max_value/(10*size))#round up to nearest 10,100,1000 etc.
        bin_ends = np.linspace(min_value,max_value,value[0]+1)
        bin_name = key+'_bin'
        data[bin_name]=pd.cut(x = data[key],
                            bins = bin_ends.tolist(), 
                            #q=value,
                            labels=False,
                            precision=0)
        if key==dep_var:
            dep_var=bin_name
        else:
            ind_vars.remove(key)
            ind_vars.append(bin_name)
        bins_str+=key+':'+str(bin_ends.tolist())+','
        bin_cts+=key+':'+str(value[0])+','
    bins_str=bins_str[:-1] 
    bin_cts=bin_cts[:-1] 
    X=data[ind_vars]
    y=pd.DataFrame(data[dep_var])
    return_dict={'X':X,'y':y,'bins_str':bins_str,'bin_cts':bin_cts}
    return(return_dict)
 
def get_model(X,y,modelType,folds):
    algos = []
    algos.append(('dtc', DecisionTreeClassifier()))
    algos.append(('rfc',RandomForestClassifier()))
    algos.append(('gbc',GradientBoostingClassifier()))
    algos.append(('rfr',RandomForestRegressor()))
    algos.append(('lr',LinearRegression()))
    algos.append(('svr',SVR()))
    results={}
    result_means={}
    if modelType=='c':
        for name, model in algos:
            result=cross_val_score(model,X,y,cv=folds,scoring='accuracy')
            results[name]=result
    elif modelType=='r':
        for name, model in algos:
            result=cross_val_score(model,X,y,cv=folds,scoring='r2')
            results[name]=result
    for key in results:
        result_means[key]=results[key].mean()
    model=min(result_means, key=result_means.get)    
    return(model)

def get_acc (X,y,ind_vars,dep_var,model,ts,folds,depth):

    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=ts,random_state=89129)
    if model=='dtc':
        algo= DecisionTreeClassifier(max_depth=depth)
    elif model=='rfc':
        algo=RandomForestClassifier(max_depth=depth)
    elif model=='gbc':
        algo=GradientBoostingClassifier(max_depth=depth)
    elif model=='rfr':
        algo=RandomForestRegressor(n_jobs=depth)
    elif model=='lr':
        algo=LinearRegression(n_jobs=depth)
    elif model=='svr':
        algo=SVR(n_jobs=depth)

    ML = algo.fit(X_train,y_train)
    preds = ML.predict(X_test)
    df_add=pd.DataFrame({'model':model,'ind_vars':[ind_vars],'dep_var':dep_var,'bins':[bins_str],'bin_cts':[bin_cts],'folds':[folds],'depth':[depth],'accuracy':[classification_report(y_test,preds,output_dict=True)['accuracy']]})
    importanceZ = zip(X.columns, ML.feature_importances_)
    importance={}
    for feat,score in importanceZ:
        importance[feat]=[score]
    return({'df':df_add,'imp':importance})
```

# Predict limit balance based on non-defaulted transactions

```python
df = pd.read_csv('credit_cleaned_transformed.csv')
df.info()
```

```python
df_acc_graph=pd.DataFrame(columns=['default','max accuracy'])
for r in range(0,2):
    data = pd.read_csv('credit_cleaned_transformed.csv')
    data=data[data['DEFAULT']==r]
    print(data.head())
    display(HTML('<h3>Default = '+str(r)+'</h3>'))
    display(HTML('<hr>'))


    classify=pd.DataFrame(columns=['model','ind_vars','dep_var','folds','depth','accuracy'])
    bins={}
    df_imp=pd.DataFrame()

    for k,d in product([5],[10]):
        try:
            ind_vars=['SEX','EDUCATION','MARRIAGE','AGE','BILL_AMT1','PAY_AMT1','PAY_AVG','PAY_0']
            for var in ind_vars:
                    if data[var].max()==data[var].min():
                        ind_vars.remove(var)
            X=data[ind_vars]
            dep_var='LIMIT_BAL'
            y=data[dep_var]
            X.head()
            model=get_model(X,y,'r',k)
            acc=get_acc(X,y,ind_vars,dep_var,model,0.3,None,None,k,d)
            add_row=acc['df']
            if df_imp.size!=0:
                df_imp=df_imp.append(pd.DataFrame.from_dict(acc['imp']))
            else:
                df_imp=pd.DataFrame.from_dict(acc['imp'])
            classify=classify.append(add_row)
        except Exception as E:
            print(E)
            continue
    cols=classify.columns.tolist()
    display(HTML('<h4>Accuracy Table<h4>'))
    display(HTML(classify[cols].sort_values(by='accuracy', ascending=False).to_html(index=False)))
    display(HTML('<h4>Bin Endpoints<h4>'))        
    #display(HTML(classify['bins'].to_frame().to_html(index=False)))
    display(HTML('<h4>Feature Importance<h4>'))
    display(HTML(df_imp.mean().to_frame().to_html(header=False)))
    df_acc_graph.loc[r-1]=[r,classify['accuracy'].max()]
ax=df_acc_graph.plot.bar(x='DEFAULT',y='max accuracy',rot=0,ylim=(0,1),legend=False)
ax.set_xlabel('DEFAULT')
ax.set_ylabel('accuracy')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
```
